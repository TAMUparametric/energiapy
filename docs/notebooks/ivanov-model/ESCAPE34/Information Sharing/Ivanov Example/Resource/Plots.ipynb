```python
import pandas
import matplotlib.pyplot as plt
from matplotlib import rc

loc1_varying_factor_ex = pandas.DataFrame(data={'com1_pur': [1, 0.58, 0.54, 0.07, 0.69, 0.17, 0.53, 0.35, 0.48, 0.39, 0.51, 0.38]})

loc6_varying_factor_ex = pandas.DataFrame(data={'com1_pur': [1, 0.26, 0.66, 0.28, 0.65, 0.01, 0.49, 0.39, 0.61, 0.44, 0.3, 0.68]})

loc1_varying_factor_ac = pandas.DataFrame(data={'com1_pur': [1, 0.58, 0.54, 0.07, 0.69, 0.17, 0.06, 0.08, 0.08, 0.05, 0.1, 0.05]})

loc6_varying_factor_ac = pandas.DataFrame(data={'com1_pur': [1, 0.26, 0.66, 0.28, 0.65, 0.01, 0.22, 0.04, 0.0, 0.1, 0.05, 0.02]})
```

```python
rc('font', **{'family': 'serif', 'serif': ['Computer Modern'], 'size': 12})
rc('text', usetex=False)

fig, ax = plt.subplots()

# Plot data from the first dataset
ax.plot(range(len(loc1_varying_factor_ex['com1_pur'])), loc1_varying_factor_ex['com1_pur'], label='Expected',
        color='blue', marker='o')

ax.plot(range(len(loc1_varying_factor_ac['com1_pur'])), loc1_varying_factor_ac['com1_pur'], label='Actual',
        color='red', marker='x')
# Customize the plot
ax.set_xlabel('Time Horizon')
ax.set_ylabel('Normalized Availability Factor')
ax.set_title('Resource Availability factor at Location 1')
ax.legend()
ax.plot(linewidth=0.5)
plt.grid(alpha=0.3)
plt.rcdefaults()
# Show the plot
plt.show()
```
