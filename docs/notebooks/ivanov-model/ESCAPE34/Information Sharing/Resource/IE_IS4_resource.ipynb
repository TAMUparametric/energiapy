```python
inventory4 = {loc1.name: {com1_store20.resource_storage.name: 0.0},
 loc2.name: {com1_store20.resource_storage.name: 20.0},
 loc3.name: {com1_store20.resource_storage.name: 0.0},
 loc4.name: {com1_store50.resource_storage.name: 0.0},
 loc5.name: {com1_store50.resource_storage.name: 0.0},
 loc6.name: {com1_store10.resource_storage.name: 0.0},
 loc7.name: {com1_store20.resource_storage.name: 5.0}}

dict_ = {i.name: {j: 0 for j in scenario_resource_IS4.set_dict['resources_store']} for i in scenario_resource_IS4.location_set}

for i in inventory4.keys():
    for j in inventory4[i].keys():
        dict_[i][j] = inventory4[i][j]

inventory4 = dict_
```

```python
problem_mincost = formulate(scenario=scenario_resource_IS4,
                            constraints={Constraints.COST, Constraints.TRANSPORT, Constraints.RESOURCE_BALANCE,
                                         Constraints.INVENTORY, Constraints.PRODUCTION,  Constraints.DEMAND},
                            demand_sign='eq', objective=Objective.COST_W_DEMAND_PENALTY, inventory_zero=inventory4)
```

```python
results_mincost = solve(scenario=scenario_resource_IS4, instance=problem_mincost, solver='gurobi', name='MILP')
```
