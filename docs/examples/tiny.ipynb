```python
from energia import *


m = Model()
m.mixed_candy = Resource(take_max=20, price=12.50)
m.blue_candy = Resource(take_max=15, price=15.67)
m.mytime = Resource(take=True, price=(7.86, 8.89))
m.sorting = Process(m.blue_candy == -0.8 * m.mixed_candy - 20*m.mytime)
```

```python
m.show()
```

```python



m.paper_mix.consume == True

m.paper_clean = Resource()
m.paper_clean.release >= 20

m.paper_waste = Resource()
m.paper_waste.release == True
m.mins = Resource()
m.mins.consume == True

m.sorting = Process()
m.sorting(m.paper_clean) == -2 * m.paper_mix + m.paper_waste - m.mins
m.sorting.capacity.x <= 200
m.sorting.operate == True

m.paper_new = Resource()
m.paper_new.consume == True
m.get_new_paper_from_printer = Process()
m.get_new_paper_from_printer.capacity.x <= 200
m.get_new_paper_from_printer.operate == True
m.get_new_paper_from_printer(m.paper_clean) == -m.paper_new - 0.5 * m.mins

m.locate(m.get_new_paper_from_printer, m.sorting)

m.mins.consume.opt()
m.paper_new.consume.opt()
```
